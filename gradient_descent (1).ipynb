## Calculating cost with gradient descent and learning rate
- Change the iteration and learning rate vaules and see the impact on cost.
- Low iteration values with high learning rate (i.e. big steps) may lead to miss the global minimum
- Goal is to reach minimum cost with minimum iteration

```python
# code credit:codebasics https://codebasics.io/coming-soon
import matplotlib.pyplot as plt
import numpy as np

learning_rate = 0.06   #change value

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 100       #change value
    n = len(x)

    cost = []

    print (len(cost))

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost.append((1/n) * sum([val**2 for val in (y-y_predicted)]))
        md = -(2/n)*sum(x*(y-y_predicted)) #-
        bd = -(2/n)*sum(y-y_predicted) #-
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {} l-rate {}, y-yp {}, x {}".format(m_curr,b_curr,cost[i], i,learning_rate,y-y_predicted, x))

    return cost

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

for i in range(7):
  learning_rate = 0.08-i*0.01   #change value
  cost = gradient_descent(x,y)
  plt.plot(cost, label=str("{0:.2f}".format(learning_rate)))
  print("Learning rate: {}".format(learning_rate))
  #plt.legend(str("{0:.2f}".format(learning_rate)))


plt.plot(cost)

plt.title("Cost function, different learning rates")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()

```
